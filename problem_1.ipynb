```python
# If you don't have kafka, please first execute pip install kafka-python
from kafka import KafkaProducer
from collections import OrderedDict
import csv
import json 
import sys

#Setting up the producer configurations
bootstrap_servers = ['localhost:9092']
topicName = 'test.topic.raw'
producer = KafkaProducer(bootstrap_servers = bootstrap_servers)
producer = KafkaProducer()

csvfile = open('partFile.csv', 'r')

reader = csv.reader(csvfile)
fieldnames = next(reader)

for row in reader:
    rowd = OrderedDict(zip(fieldnames, row))
    data = json.dumps(rowd)
    print(data)
    producer.send(topicName, value=data.encode('utf-8'))
```
