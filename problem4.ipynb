```python
from kafka import KafkaConsumer
from kafka import KafkaProducer
import sys
import json
import os

### Setting up the Python consumer
bootstrap_servers = ['localhost:9092']
topicName= 'test.topic.raw'
consumer = KafkaConsumer (topicName, group_id = 'sales',bootstrap_servers = bootstrap_servers, auto_offset_reset = 'latest')   ## You can also set it as latest
#consumer2 = KafkaConsumer (topicName, group_id = 'def',bootstrap_servers = bootstrap_servers, auto_offset_reset = 'earliest')   ## You can also set it as latest

                           
server_topics = consumer.topics()

#Setting up the producer configurations
producer = KafkaProducer(bootstrap_servers=bootstrap_servers)
producer = KafkaProducer()

### Reading the message from consumer
try:
    for msg in consumer :
        
        row = json.loads(msg.value)
       
        print(msg.value)
        
        sales_ch = row["Sales Channel"]
        topic1 = 'test.topic.' + sales_ch
        print(topic1)
        
        unit_cost = row["Unit Cost"]
        topic2 = 'topic.costly_order'
        print(topic2)
        
        if (topic1 not in server_topics):
            os.system("downloads/kafka_2.12-2.3.0/bin/kafka-topics.sh --create --bootstrap-server localhost:9092 --replication-factor 1 --partitions 1 --topic " + topic1)
            producer.send(topic1, msg.value)
        else:
            producer.send(topic1, msg.value)
                           
        
        
        if (float(unit_cost) > 10000):
            if (topic2 not in server_topics):
                os.system("downloads/kafka_2.12-2.3.0/bin/kafka-topics.sh --create --bootstrap-server localhost:9092 --replication-factor 1 --partitions 1 --topic " + topic2)
                producer.send(topic2, msg.value)
            else:
                producer.send(topic2, msg.value)
except KeyboardInterrupt:
    sys.exit()
```
