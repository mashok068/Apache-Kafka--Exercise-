```python
from kafka import KafkaConsumer
from kafka import KafkaProducer
import sys
import json 
import os
from kafka.admin import KafkaAdminClient, NewTopic


admin_client = KafkaAdminClient(bootstrap_servers="localhost:9092", client_id='test')

topic_list = []


### Setting up the Python consumer
bootstrap_servers = ['localhost:9092']
topicName= 'test.topic.raw'
consumer = KafkaConsumer (topicName, group_id = 'by_country',bootstrap_servers = bootstrap_servers, auto_offset_reset = 'latest')   ## You can also set it as latest
server_topics = consumer.topics()

#print(server_topics)

#Setting up the producer configurations
producer = KafkaProducer(bootstrap_servers = bootstrap_servers)
producer = KafkaProducer()

### Reading the message from consumer
try:
    for message in consumer:
        row = json.loads(message.value)
        # Sending the message with both key and value
        print(message.value)
        country = row["Country"].replace(" ", "_")
        
        topic = 'test.topic.' + country
        print(topic)
        
        if (topic not in server_topics):
            os.system("downloads/kafka_2.12-2.3.0/bin/kafka-topics.sh --create --bootstrap-server localhost:9092 --replication-factor 1 --partitions 1 --topic " + topic)
            producer.send(topic, message.value)
        else:
            producer.send(topic, message.value)
        
except KeyboardInterrupt:
    sys.exit()
```
